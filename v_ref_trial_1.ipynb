```python
%%time
import warnings
warnings.filterwarnings("ignore")
import pyart
import wradlib as wrl
import numpy as np
import xarray as xr
from matplotlib import pyplot as plt
import glob, os, sys
%matplotlib inline
```

```python
ds = xr.open_dataset("../IMD_Radar_stuff/PyScanCf/Examples/goa16/polar_GOA210516024101.nc")
```

```python
ax, im = wrl.vis.plot_ppi(ds['REF'][0:360], r = ds.range/1e3, cmap="pyart_NWSRef", vmin = -10, vmax = 70,
                 az = ds.azimuth[0:360].data, proj="cg")
plt.colorbar(im)
```

```python
dsnew = ds.swap_dims({"time":"elevation"})
```

```python
ds11 = dsnew.where((dsnew.azimuth >=270)&((dsnew.azimuth <=271)), drop=True)
```

```python
ds11['REF'].plot(vmin = -10, vmax = 70, figsize = [10, 4], cmap = "pyart_NWSRef")
```

```python
ds11 = dsnew.where((dsnew.azimuth >=250)&((dsnew.azimuth <=265)), drop=True)
ds11['REF'].plot(vmin = -10, vmax = 70, figsize = [10, 4], cmap = "pyart_NWSRef")
```
